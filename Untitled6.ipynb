```python
# Import required libraries for analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Improve default plot appearance
plt.style.use("ggplot")
```

```python
# Load cleaned Aadhaar enrolment dataset
enrol = pd.read_csv("api_data_aadhar_enrolment_cleaned.csv")

# Load cleaned demographic dataset
demo = pd.read_csv("api_data_aadhar_demographic_cleaned.csv")

# Load cleaned biometric dataset
bio = pd.read_csv("api_data_aadhar_biometric_cleaned.csv")

# Quick sanity check
print(enrol.shape, demo.shape, bio.shape)
```

```python
# Check column names to ensure correct loading
print(enrol.columns)
print(demo.columns)
print(bio.columns)
```

```python
# Ensure required columns for temporal analysis are present
required_cols = ['year', 'month', 'total_enrolment']
assert all(col in enrol.columns for col in required_cols), "Missing required columns"

print("All required columns are present.")
```

```python
# Aggregate total Aadhaar enrolment year-wise
yearly_enrolment = (
    enrol
    .groupby('year')['total_enrolment']
    .sum()
    .reset_index()
    .sort_values('year')
)

yearly_enrolment
```

```python
# Plot year-wise Aadhaar enrolment trend
plt.figure(figsize=(10,5))

plt.plot(
    yearly_enrolment['year'],
    yearly_enrolment['total_enrolment'],
    marker='o'
)

plt.title("Year-wise Aadhaar Enrolment Trend")
plt.xlabel("Year")
plt.ylabel("Total Enrolment")
plt.tight_layout()
plt.show()
```

```python
# Plot year-wise Aadhaar enrolment trend
plt.figure(figsize=(10,5))

plt.plot(
    yearly_enrolment['year'],
    yearly_enrolment['total_enrolment'],
    marker='o'
)

plt.title("Year-wise Aadhaar Enrolment Trend")
plt.xlabel("Year")
plt.ylabel("Total Enrolment")
plt.tight_layout()
plt.show()
```

```python
# Aggregate enrolment month-wise across all years
monthly_enrolment = (
    enrol
    .groupby('month')['total_enrolment']
    .sum()
    .reset_index()
    .sort_values('month')
)

monthly_enrolment
```

```python
# Plot month-wise Aadhaar enrolment pattern
plt.figure(figsize=(10,5))

plt.bar(
    monthly_enrolment['month'],
    monthly_enrolment['total_enrolment']
)

plt.title("Month-wise Aadhaar Enrolment Pattern")
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.tight_layout()
plt.show()
```

```python
# Plot month-wise Aadhaar enrolment pattern
plt.figure(figsize=(10,5))

plt.bar(
    monthly_enrolment['month'],
    monthly_enrolment['total_enrolment']
)

plt.title("Month-wise Aadhaar Enrolment Pattern")
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.tight_layout()
plt.show()
```

```python
# Aggregate total enrolment at state level
state_enrolment = (
    enrol
    .groupby('state')['total_enrolment']
    .sum()
    .reset_index()
    .sort_values('total_enrolment', ascending=False)
)

state_enrolment.head()
```

```python
# Select top 10 states with highest enrolment
top_states = state_enrolment.head(10)

plt.figure(figsize=(10,6))
plt.barh(
    top_states['state'],
    top_states['total_enrolment']
)
plt.gca().invert_yaxis()  # Highest at top

plt.title("Top 10 States by Aadhaar Enrolment")
plt.xlabel("Total Enrolment")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
# Select bottom 10 states with lowest enrolment
bottom_states = state_enrolment.tail(10)

plt.figure(figsize=(10,6))
plt.barh(
    bottom_states['state'],
    bottom_states['total_enrolment']
)
plt.gca().invert_yaxis()

plt.title("Bottom 10 States by Aadhaar Enrolment")
plt.xlabel("Total Enrolment")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
# Aggregate enrolment at district level
district_enrolment = (
    enrol
    .groupby(['state', 'district'])['total_enrolment']
    .sum()
    .reset_index()
    .sort_values('total_enrolment', ascending=False)
)

district_enrolment.head()
```

```python
# Select top 10 districts
top_districts = district_enrolment.head(10)

plt.figure(figsize=(10,6))
plt.barh(
    top_districts['district'],
    top_districts['total_enrolment']
)
plt.gca().invert_yaxis()

plt.title("Top 10 Districts by Aadhaar Enrolment")
plt.xlabel("Total Enrolment")
plt.ylabel("District")
plt.tight_layout()
plt.show()
```

```python
# Ensure required columns for demographic analysis are present
required_cols = [
    'demo_age_5_17',
    'demo_age_17_',
    'total_demographic_activity'
]

assert all(col in demo.columns for col in required_cols), "Missing required demographic columns"

print("All required demographic columns are present.")
```

```python
# Calculate total Aadhaar activity for each age group
age_group_totals = {
    'Age 517': demo['demo_age_5_17'].sum(),
    'Age 17+': demo['demo_age_17_'].sum()
}

age_group_df = pd.DataFrame(
    list(age_group_totals.items()),
    columns=['Age Group', 'Total Activity']
)

age_group_df
```

```python
# Plot demographic participation by age group
plt.figure(figsize=(7,5))

plt.bar(
    age_group_df['Age Group'],
    age_group_df['Total Activity']
)

plt.title("Aadhaar Participation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Activity")
plt.tight_layout()
plt.show()
```

```python
# Aggregate demographic activity by state and age group
state_demo = (
    demo
    .groupby('state')[['demo_age_5_17', 'demo_age_17_']]
    .sum()
    .reset_index()
)

state_demo.head()
```

```python
# Select top 10 states by total demographic activity
state_demo['total'] = state_demo['demo_age_5_17'] + state_demo['demo_age_17_']
top_states_demo = state_demo.sort_values('total', ascending=False).head(10)

# Plot stacked bar chart
plt.figure(figsize=(12,6))

plt.bar(
    top_states_demo['state'],
    top_states_demo['demo_age_5_17'],
    label='Age 517'
)

plt.bar(
    top_states_demo['state'],
    top_states_demo['demo_age_17_'],
    bottom=top_states_demo['demo_age_5_17'],
    label='Age 17+'
)

plt.xticks(rotation=45, ha='right')
plt.title("Age-wise Aadhaar Participation in Top 10 States")
plt.xlabel("State")
plt.ylabel("Total Activity")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Aggregate biometric activity month-wise across all regions
monthly_bio = (
    bio
    .groupby(['year', 'month'])['total_biometric_activity']
    .sum()
    .reset_index()
    .sort_values(['year', 'month'])
)

monthly_bio.head()
```

```python
# Create a time index for better plotting
monthly_bio['time'] = pd.to_datetime(
    monthly_bio['year'].astype(str) + '-' +
    monthly_bio['month'].astype(str) + '-01'
)

plt.figure(figsize=(12,5))

plt.plot(
    monthly_bio['time'],
    monthly_bio['total_biometric_activity'],
    marker='o'
)

plt.title("Monthly Biometric Activity Trend")
plt.xlabel("Time")
plt.ylabel("Total Biometric Activity")
plt.tight_layout()
plt.show()
```

```python
# Calculate mean and standard deviation
mean_bio = monthly_bio['total_biometric_activity'].mean()
std_bio = monthly_bio['total_biometric_activity'].std()

# Compute Z-score
monthly_bio['z_score'] = (
    (monthly_bio['total_biometric_activity'] - mean_bio) / std_bio
)

monthly_bio.head()
```

```python
# Flag anomalies where deviation is significant
anomalies = monthly_bio[monthly_bio['z_score'].abs() > 2]

anomalies
```

```python
plt.figure(figsize=(12,5))

# Plot normal trend
plt.plot(
    monthly_bio['time'],
    monthly_bio['total_biometric_activity'],
    label='Biometric Activity',
    marker='o'
)

# Highlight anomalies
plt.scatter(
    anomalies['time'],
    anomalies['total_biometric_activity'],
    color='red',
    label='Anomalies',
    zorder=5
)

plt.title("Biometric Activity with Detected Anomalies")
plt.xlabel("Time")
plt.ylabel("Total Biometric Activity")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Aggregate biometric activity at state level
state_bio = (
    bio
    .groupby('state')['total_biometric_activity']
    .sum()
    .reset_index()
    .sort_values('total_biometric_activity', ascending=False)
)

state_bio.head()
```

```python
# Aggregate total enrolment by state
state_enrol = (
    enrol
    .groupby('state')['total_enrolment']
    .sum()
    .reset_index()
)
```

```python
# Aggregate total enrolment by state
state_enrol = (
    enrol
    .groupby('state')['total_enrolment']
    .sum()
    .reset_index()
)
```

```python
# Aggregate total biometric activity by state
state_bio = (
    bio
    .groupby('state')['total_biometric_activity']
    .sum()
    .reset_index()
)
```

```python
# Merge all three datasets on state
state_combined = (
    state_enrol
    .merge(state_demo, on='state', how='inner')
    .merge(state_bio, on='state', how='inner')
)

state_combined.head()
```

```python
# Biometric load relative to enrolment
state_combined['bio_to_enrol_ratio'] = (
    state_combined['total_biometric_activity'] /
    state_combined['total_enrolment']
)

# Biometric load relative to demographic activity
state_combined['bio_to_demo_ratio'] = (
    state_combined['total_biometric_activity'] /
    state_combined['total']
)

state_combined.head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    state_combined['total_enrolment'],
    state_combined['total_biometric_activity']
)

plt.xlabel("Total Enrolment")
plt.ylabel("Total Biometric Activity")
plt.title("Enrolment vs Biometric Activity (State Level)")
plt.tight_layout()
plt.show()
```

```python
# Sort states by biometric-to-enrolment ratio
mismatch_states = (
    state_combined
    .sort_values('bio_to_enrol_ratio', ascending=False)
)

mismatch_states[['state', 'bio_to_enrol_ratio']].head(10)
```

# **Key Insights from Analysis**

Aadhaar enrolment activity shows clear temporal patterns, with identifiable long-term trends and seasonal peaks, indicating predictable demand cycles.

Enrolment and demographic participation are unevenly distributed across regions, with a small number of states and districts accounting for a large share of activity, while several regions remain underserved.

Adult population groups (17+) contribute the majority of Aadhaar-related activity, while participation among younger age groups (517) is comparatively lower across most regions.

Biometric activity exhibits periodic anomalies, suggesting episodes of operational stress, special enrolment drives, or system-level events.

In several states, biometric activity is not proportionate to enrolment and demographic participation, indicating mismatches between service demand and operational capacity.

# **Recommendations for UIDAI Service Delivery Planning**

1. Demand-Based Infrastructure Planning

Regions with consistently high enrolment and biometric activity should be prioritized for:

Additional enrolment and update centres

Increased staffing during peak periods

Infrastructure upgrades to handle sustained demand

This ensures service availability aligns with actual usage patterns

2. Targeted Outreach for Underserved Regions

Regions with low enrolment and demographic participation should be:

Targeted with awareness campaigns

Supported through mobile enrolment units

Monitored for access-related constraints

This can help improve Aadhaar coverage and inclusion.

3. Age-Specific Engagement Strategies

Lower participation among younger age groups suggests the need for:

School- and community-based enrolment drives

Simplified communication tailored to parents and guardians

This supports balanced demographic coverage over time.

4. Proactive Monitoring of Biometric Anomalies

Periods of abnormal biometric activity should trigger:

Early operational alerts

Temporary capacity augmentation

Root-cause analysis to distinguish between campaigns and system issues

This helps maintain service reliability and reduce downtime.

5. Rebalancing Biometric Resources

States showing high biometric load relative to enrolment should be reviewed for:

Biometric device availability

System throughput constraints

Process optimization opportunities

Conversely, under-utilized regions can be optimized to improve efficiency.
